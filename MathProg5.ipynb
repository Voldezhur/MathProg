Пример с СДО

```python
import matplotlib.pyplot as plt
import numpy as np
def draw_triangle(vertices, level, ax):
 if level == 0:
  triangle = plt.Polygon(vertices, edgecolor='black')
  ax.add_patch(triangle)
 else:
   # Находим середины сторон
  midpoints = [(vertices[i] + vertices[(i+1) % 3]) / 2 for i in range(3)]

  # Рекурсивно вызываем для каждого подмножества вершин
  draw_triangle([vertices[0], midpoints[0], midpoints[2]], level-1, ax)
  draw_triangle([vertices[1], midpoints[0], midpoints[1]], level-1, ax)
  draw_triangle([vertices[2], midpoints[1], midpoints[2]], level-1, ax)

# Основные вершины большого треугольника
vertices = np.array([[0, 0], [1, np.sqrt(3)], [2, 0]])
fig, ax = plt.subplots()
ax.set_aspect('equal')
# Построение фрактала с 4 уровнями рекурсии
draw_triangle(vertices, 4, ax)
# Настройка отображения
ax.set_xlim([-0.1, 2.1])
ax.set_ylim([-0.1, np.sqrt(3)+0.1])
plt.axis('off')
plt.show()
```

Задача 1 - Мандельбротов факториал

```python
import matplotlib.pyplot as plt
import numpy as np

def complex_matrix(xmin, xmax, ymin, ymax, pixel_density):
    re = np.linspace(xmin, xmax, int((xmax - xmin) * pixel_density))
    im = np.linspace(ymin, ymax, int((ymax - ymin) * pixel_density))
    return re[np.newaxis, :] + im[:, np.newaxis] * 1j

def is_stable(c, num_iterations):
    z = 0
    for _ in range(num_iterations):
        z = z ** 2 + c
    return abs(z) <= 2


c = complex_matrix(-1.5, -1, -0.2, 0.2, pixel_density=2048)  # Значения по умолчанию

plt.imshow(is_stable(c, num_iterations=20), cmap="binary")
plt.gca().set_aspect("equal")
plt.axis("off")
plt.tight_layout()
plt.show()
```
